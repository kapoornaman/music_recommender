```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("songdata.csv")
```

```python
print(df.head())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
songs=df.sample(n=30000).drop("link",axis=1).reset_index(drop=True)
```

```python
songs["text"]=songs["text"].str.replace('\n','' )
```

```python
songs.head()
```

```python
tfidf=TfidfVectorizer(analyzer="word",stop_words="english")
lyrics_matrix=tfidf.fit_transform(songs["text"])
```

```python
similarities={}
```

```python
cosine_similarities=cosine_similarity(lyrics_matrix)
```

```python
for i in range(len(cosine_similarities)):
    similar_indicies=cosine_similarities[i].argsort()[:-50:-1]
    similarities[songs['song'].iloc[i]] = [(cosine_similarities[i][x], songs['song'][x], songs['artist'][x]) for x in similar_indicies][1:]
    
```

```python
class ContentBasedRecommender:
    def __init__(self, matrix):
        self.matrix_similar = matrix

    def _print_message(self, song, recom_song):
        rec_items = len(recom_song)
        for i in range(rec_items):
            print(f"{recom_song[i][1]} by {recom_song[i][2]} with {round(recom_song[i][0], 3)} similarity score") 
    def recommend(self, recommendation):
        # Get song to find recommendations for
        song = recommendation['song']
        # Get number of songs to recommend
        number_songs = recommendation['number_songs']
        print(f"For {song} the closest are \n")
        # Get the number of songs most similars from matrix similarities
        recom_song = self.matrix_similar[song][:number_songs]
        # print each item
        self._print_message(song=song, recom_song=recom_song)
```

```python
recommendations=ContentBasedRecommender(similarities)
```

```python
recommendations2={
    "song": songs['song'].iloc[9997],
    "number_songs":5    
}
```

```python
recommendations.recommend(recommendations2)
```
